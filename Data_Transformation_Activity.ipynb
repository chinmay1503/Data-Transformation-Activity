```python
csv_dropped_columns = df = pd.read_csv(r'.\books.csv', usecols=['Identifier', 'Place of Publication', 'Date of Publication', 'Publisher', 'Title', 'Author', 'Contributors', 'Flickr URL'])
```

```python
# Remove the extra dates in square brackets, wherever present: e.g., 1879 [1878] should be converted to 1879
csv_dropped_columns['Date of Publication'] = csv_dropped_columns['Date of Publication'].replace(to_replace='\[\d*]', value="", regex=True)
csv_dropped_columns['Date of Publication'] = csv_dropped_columns['Date of Publication'].str.strip()
```

```python
# Convert date ranges to their “start date”: e.g., 1860-63; 1839, 38-54
csv_dropped_columns['Date of Publication'] = csv_dropped_columns['Date of Publication'].replace(to_replace='\-\d*', value=np.NAN, regex=True)
csv_dropped_columns['Date of Publication']
```

```python
# Remove uncertain dates and replace them with NumPy’s NaN: [1897?], #Convert the string nan to NumPy’s NaN value
csv_dropped_columns['Date of Publication'] = csv_dropped_columns['Date of Publication'].replace(to_replace='.{5,}', value=np.NAN, regex=True)
csv_dropped_columns['Date of Publication']
```

```python
csv_dropped_columns['Date of Publication'] = pd.to_numeric(csv_dropped_columns['Date of Publication'], errors='coerce')
```

```python
uniplacesDF = pd.read_csv(r'.\uniplaces.txt', sep=";", names=['Region Name'])

uniplacesDF['State']= uniplacesDF['Region Name'].str.extract('(.*)\[edit\]', expand=False).ffill()
uniplacesDF['Town']= uniplacesDF['Region Name'].str.extract('(.*)\(\D*\)', expand=False).ffill()
uniplacesDF['University']= uniplacesDF['Region Name'].str.extract('\((.*?)\)', expand=False).ffill()

uniplacesDF = uniplacesDF.drop('Region Name', axis=1)
uniplacesDF = uniplacesDF.dropna()
import re

def cleanColumns(val):
         val = re.sub(r'\(.*', "", str(val))
         val = re.sub(r'-.*', "", str(val))
         val = re.sub(r'–.*', "", str(val))
         val = re.sub(r'\[\d*]', "", str(val))
         print(val)
         return val;

uniplacesDF = uniplacesDF.applymap(cleanColumns)

uniplacesDF.to_csv('divided.csv')
```
